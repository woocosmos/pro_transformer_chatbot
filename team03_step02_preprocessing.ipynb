```python
import pandas as pd
```

```python
data1 = pd.read_excel('/content/B 의류(15,826).xlsx') # 의류
data2 = pd.read_excel('/content/H 관광여가오락(4,949).xlsx') # 관광
data3 = pd.read_excel('/content/차량등록_최종본(0429).xlsx') # 차량
```

# 의류

```python
test = data1[['SPEAKER', 'SENTENCE']]
test = test[:14900]
```

```python
test['match'] = test.SPEAKER.eq(test.SPEAKER.shift()) # 화자 중복 여부
test['SUB'] = test.SENTENCE.shift(-1) # 대사 1행씩 땡기기
test["FULL"] = test["SENTENCE"].astype(str) + " " + test["SUB"].astype(str) # 원래 대사와 이후 대사 합치기
```

```python
idx = test[test['match']==True].index # 화자가 중복되는 인덱스
idx
```

```python
idx = idx-1
for i in range(len(test)):
    if i in idx:
        test.loc[i, 'SENTENCE'] = test.loc[i].FULL # 화자가 중복되면 이전 대화+이후 대사로 교체

test.drop(idx+1, inplace=True) # 이후 대사 삭제
test = test[['SPEAKER', 'SENTENCE']]
test.reset_index(drop=True, inplace=True)
```

```python
Q_sample_data =  test[test['SPEAKER']=='고객']
Q_data = Q_sample_data['SENTENCE']
A_sample_data =  test[test['SPEAKER']=='점원']
A_data = A_sample_data['SENTENCE']
Q_data = Q_data.reset_index()
A_data = A_data.reset_index()
del Q_data['index']
del A_data['index']
QA_Data = pd.concat([Q_data,A_data],axis=1)
QA_Data.columns = ['Q','A']
QA_Data
```

```python
QA_Data.to_excel('의류.xlsx')
```
